```python
# load one wind profile
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# load the data from csv
data = pd.read_csv('wind-profiles/CR2125/profile_wso_CR2125_line_0001.csv', delimiter=',', header=0, names=['R', 'L', 'Lon', 'Lat', 'n', 'v', 'T', 'B', 'A/A0', 'alpha', 'V/Cs', 'propag_dt'], skiprows=2)
data.head()
```

```python
# plot the wind profile
variables = ['R', 'L', 'Lon', 'Lat', 'n', 'v', 'T', 'B', 'A/A0', 'alpha', 'V/Cs', 'propag_dt']
data[variables].plot(x='R', subplots=True, layout=(4, 3), figsize=(12, 8))
plt.tight_layout()
plt.show()
```

```python
# see which number is missing in CR1992 folder
# profile_wso_CR1992_line_xxxx.csv where xxxx is the number of the profile

import os
import re

# get the list of files in the folder
files = os.listdir('wind-profiles/CR1992')
# get the list of numbers in the file names profile_wso_CR1992_line_xxxx.csv
numbers = [int(re.search('profile_wso_CR1992_line_([0-9]+).csv', file).group(1)) for file in files]
# find the missing number
missing = [i for i in range(1, max(numbers)) if i not in numbers]
print(missing)
```
